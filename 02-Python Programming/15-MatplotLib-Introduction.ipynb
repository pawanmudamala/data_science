### B - Static Graphics

```python
%matplotlib inline
```

```python
x = [1,2,3,4]
y = [1,4,9,16]

plt.plot(x,y);
```

```python
plt.plot(x, y, color='green', marker='o', linestyle='dashed',linewidth=2, markersize=5)
```

```python
plt.plot(x, y, 'go--', linewidth=2, markersize=12)
```

```python
d = np.arange(0.,5.,0.2)
d
```

```python
len(d)
```

```python
plt.plot(d, d,'r--',d,d**2,'bs',d,d**3,'g^');
```
