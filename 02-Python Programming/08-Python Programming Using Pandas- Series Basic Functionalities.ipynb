```python
import pandas as pd
```

```python
data={"name":["Bill","Tom","Tim","John","Alex","Vanessa","Kate"],      
      "score":[90,80,85,75,95,60,65],      
      "sport":["Wrestling","Football","Skiing","Swimming","Tennis",
               "Karete","Surfing"],      
      "sex":["M","M","M","M","F","F","F"]}

df = pd.DataFrame(data)
df
```

```python
name_series = df['name']
name_series
```

### 1 - axes - Returns a list of row axis labels

```python
name_series.axes
```

```python
index_list = []
for i in range(0,df.shape[0]) :
    index_list.append("Index_"+str(i))
index_list
```

```python
name_series.index = index_list
```

```python
name_series.axes
```

```python
df
```

### 2 - dtype - Returns the type of the object

#### - Name Series contains only the strings so the data type is Object

```python
name_series.dtype
```

#### - Score Series contains only the integers so the data type is int64

```python
df['score'].dtype
```

### 3 - empty - Returns TRUE is the series is empty

```python
name_series.empty
```

```python
empty_series = pd.Series()
empty_series.empty
```

### 4 - ndim - Returns the number of dimensions of the underlying data

```python
name_series.ndim
```

### 5 - size - Returns the size(length) of the series.

```python
name_series.size
```

```python
df.shape
```

### 6 - values - Returns the actual data in the series as an array.

```python
name_series.values
```

```python
names = name_series.values
type(names)
```
