## Scatter Plot with Scatter Method

```python
plt.scatter(iris['sepal.length'],iris['sepal.width'])
```

```python
plt.scatter(iris['sepal.length'],iris['sepal.width'],s=75)
```

```python
plt.scatter(iris['sepal.length'],iris['sepal.width'], s=75, c="m")
```

```python
plt.scatter(iris['sepal.length'],iris['sepal.width'], s=75, c="m", marker="v")
```

```python
plt.scatter(iris['sepal.length'],iris['sepal.width'], 
            s=75, c="m", 
            edgecolor="black", 
            linewidth=3)
```

```python
plt.scatter(iris['sepal.length'],iris['sepal.width'],
            s=75, c="m", 
            edgecolor="black", 
            linewidth=3, 
            alpha=0.5)
```

```python
color = np.random.randint(0,10,150)
color
```

```python
plt.scatter(iris['sepal.length'],iris['sepal.width'], 
            s=100, 
            c=color, 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
```

```python
plt.scatter(iris['sepal.length'],iris['sepal.width'], 
            s=100, 
            c=color, 
            cmap="Reds", 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
```

```python
plt.scatter(iris['sepal.length'],iris['sepal.width'], 
            s=100, 
            c=color, 
            cmap="viridis", 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
```

```python
plt.scatter(iris['sepal.length'],iris['sepal.width'], 
            s=100, 
            c=color, 
            cmap="viridis", 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
cbar=plt.colorbar()
```

```python
plt.scatter(iris['sepal.length'],iris['sepal.width'], 
            s=100, 
            c=color, 
            cmap="viridis", 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
cbar=plt.colorbar()
cbar.set_label("Levels")
```

```python
sizes = np.random.randint(50,200,150)
sizes
```

```python
plt.scatter(iris['sepal.length'],iris['sepal.width'], 
            s=sizes,  
            c=color, 
            cmap="Reds",
            edgecolor="black",
            linewidth=1, 
            alpha=1)
```

```python
plt.scatter(iris['sepal.length'],iris['sepal.width'], 
            s=sizes,  
            c=color, 
            cmap="Reds",
            edgecolor="black",
            linewidth=1, 
            alpha=1)
plt.xlabel("sepal_width")
plt.ylabel("petal_length")
```

```python
plt.scatter(iris['sepal.length'],iris['sepal.width'], 
            s=sizes,  
            c=color, 
            cmap="Reds",
            edgecolor="black",
            linewidth=1, 
            alpha=1)
plt.xlabel("sepal_width")
plt.ylabel("petal_length")
plt.title("İris Sepal Length & Sepal Width ")
```
