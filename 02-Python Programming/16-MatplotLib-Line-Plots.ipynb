# Line Plots in Python using Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

## Creating a Graphic Object 

```python
fig = plt.figure()
ax = plt.axes()
```

```python
x = np.linspace(0,10)
```

```python
y = np.sin(x)
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x,y)
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

## Specifying Line Colors

```python
plt.plot(x, np.sin(x - 0), color='blue') 
```

```python
plt.plot(x, np.sin(x - 1), color='g')
```

```python
plt.plot(x, np.sin(x - 2), color='0.75') 
```

```python
plt.plot(x, np.sin(x - 3), color='#FFDD44')
```

## Specifying the Line Style

```python
plt.plot(x, x + 0, linestyle='solid')
```

```python
plt.plot(x, x + 1, linestyle='dashed')
```

```python
plt.plot(x, x + 2, linestyle='dashdot')
```

Grafii noktalar ile gsterelim.

```python
plt.plot(x, x + 3, linestyle='dotted')
```

```python
plt.plot(x, x + 4, linestyle='-') 
plt.plot(x, x + 5, linestyle='--') 
plt.plot(x, x + 6, linestyle='-.') 
plt.plot(x, x + 7, linestyle=':');
```

```python
plt.plot(x, x + 0, '-r')
```

```python
plt.plot(x, x + 1, '--b')
```

```python
plt.plot(x, x + 2, '-.g') 
```

```python
plt.plot(x, x + 3, ':k')
```

## Specifying the Values of the Axes

```python
plt.plot(x, np.sin(x))
plt.xlim(-5, 20)
plt.ylim(-3, 3)
```

```python
plt.plot(x, np.sin(x))
plt.xlim(-5, 20)
plt.ylim(3,-3)
```

```python
plt.plot(x, np.sin(x))
plt.axis([-5, 10, -3, 3])
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
```

## Adding a Label to a Chart

```python
plt.plot(x, np.sin(x))
plt.title("A sine curve")
```

```python
plt.plot(x, np.sin(x))
plt.title("A sine curve")
plt.xlabel("x values")
plt.ylabel("sin(x) values")
```

```python
plt.plot(x, x+2, '-r', label='red')
plt.plot(x, x+4, ':b', label='blue')
plt.axis('equal')
plt.legend()
```
